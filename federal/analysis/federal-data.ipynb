```python
def vdc_to_geocode(district, vdc):
    district_code = get_district_geocode(district)
    vdc = ' '.join(vdc.split()[:-1])
    vdc_name = mapping.get(vdc) if mapping.get(vdc) else vdc
    vdc_geo = dict(vdc_df[vdc_df.parent_code == district_code][['name', 'geo_code']].values)
    return vdc_geo[vdc_name]
```

```python
headers = ['geo_level', 'geo_code', 'sex', 'total']
final_df = pd.DataFrame(columns=headers)

# Data for each local level
for province in provinces:
    districts = os.listdir(os.path.join(folder, province))
    for district in districts:
        vdcs = os.listdir(os.path.join(folder, province, district))
        for vdc in vdcs:
            try:
                df = pd.read_csv(os.path.join(folder, province, district, vdc))
            except Exception:
                print(vdc, district, province)
            if 'Total Female' in df.columns:
                col = 'Total Female'
            elif 'Total Female ' in df.columns:
                col = 'Total Female '
            else:
                print(vdc, district, province)
                raise ValueError()
            df = df.rename({'Total Male': 'male', col: 'female'}, axis=1)
            try:
                df = pd.melt(df[df['Ward No.'] == 'Total'][['male', 'female']], var_name='sex', value_name='total')
            except Exception:
                print(vdc, district, province)
            df['geo_level'] = 'local'
            try:
                df['geo_code'] = vdc_to_geocode(district, vdc)
            except Exception:
                # Data inconsistency for vdc level
                continue
            df = df[['geo_level', 'geo_code', 'sex', 'total']]
            final_df = pd.concat([final_df, df])
```

```python
# Aggregation of local data for district level
district_codes = geo_df[geo_df.geo_level == 'district']['geo_code'].values
for district_code in district_codes:
    local_codes = geo_df[geo_df.parent_code == district_code].geo_code.values
    district_value = final_df[final_df.geo_code.isin(local_codes)].groupby('sex')['total'].sum()
    try:
        district_data = [{'geo_level': 'district', 'geo_code': district_code, 'total': district_value['male'], 'sex': 'male'},
                     {'geo_level': 'district', 'geo_code': district_code, 'total': district_value['female'], 'sex': 'female'}]
    except Exception:
        # District present in geo but not in federal data
        # Skipping for now
        print(district)
        continue
    district_df = pd.DataFrame(district_data)
    final_df = final_df.append(district_df)
```

```python
# Aggregation of all data for country level
total_male = final_df.groupby('sex')['total'].sum()['male']
total_female = final_df.groupby('sex')['total'].sum()['female']
data = [{'total': total_male, 'sex': 'male', 'geo_level': 'country', 'geo_code': 'NP'},
        {'total': total_female, 'sex': 'female', 'geo_level': 'country', 'geo_code': 'NP'}]
country_df = pd.DataFrame(data, columns=['geo_level', 'geo_code', 'sex', 'total'])
final_df = final_df.append(country_df)
```

```python
# Aggregation of data for province level

# Create province to district mapping
codes = geo_df[geo_df.geo_level == 'district'][['parent_code', 'geo_code']].values
province_districts = {}
for province, district in codes:
    if province in province_districts:
        province_districts[province].append(district)
    else:
        province_districts[province] = [district]
        
# Aggregate and add data of province to dataframe
for province, districts in province_districts.items():
    province_df = final_df[final_df.geo_code.isin(districts)]
    population_count = province_df.groupby('sex')['total'].sum()
    data = [{'total': population_count['male'], 'sex': 'male', 'geo_level': 'province', 'geo_code': province},
        {'total':  population_count['male'], 'sex': 'female', 'geo_level': 'province', 'geo_code': province}]
    province_df = pd.DataFrame(data)
    final_df = final_df.append(province_df)
```

```python
# Generate final CSV. Now convert to postgres sql.
final_df.to_csv('population_data.csv', index=False)
```
